```python
# Importações

import pytesseract
import cv2
from google.colab.patches import cv2_imshow
from google.colab import drive
```

```python
# Montando o drive
drive.mount('/content/drive')
```

```python
# Carregando e pré processando a imagem

alpha = 1 # Controle de contraste
beta = -0.5 # Controle de luminosidade

img = cv2.imread('/content/drive/MyDrive/Arquivos - Dados/Curso - visão computacional/Images/trecho-livro.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)
cv2_imshow(img)
```

```python
!tesseract --help-psm
```

```python
#Conversão de imagem para texto
config_tesseract = '--tessdata-dir tessdata --psm 6'
text = pytesseract.image_to_string(img, lang = 'por', config = config_tesseract)
print(text)
```
