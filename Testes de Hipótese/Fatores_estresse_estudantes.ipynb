#Análise Comparativa:

1. Existe uma correlação entre o nível de ansiedade e o desempenho acadêmico?

2. Alunos com má qualidade de sono também relatam níveis mais altos de depressão?

3. Alunos que sofrem bullying são mais propensos a ter histórico de problemas de saúde mental?

```python
corr_ansiedade_academica = df.anxiety_level.corr(df.academic_performance)

corr_abs = abs(corr_ansiedade_academica)

if corr_abs < 0.3:
    print('Não há correlação entre as duas variáveis')
elif 0.3 <= corr_abs < 0.5:
    print(f'A correlação de Pearson é fraca, com valor de {round(corr_ansiedade_academica, 2)}')
elif 0.5 <= corr_abs < 0.7:
    print(f'A correlação de Pearson é moderada, com valor de {round(corr_ansiedade_academica, 2)}')
elif corr_abs >= 0.7:
    print(f'A correlação de Pearson é forte, com valor de {round(corr_ansiedade_academica, 2)}')
```

```python
corr_depressao_sono = df.depression.corr(df.anxiety_level)

corr_abs = abs(corr_depressao_sono)

if corr_abs < 0.3:
    print('Não há correlação entre as duas variáveis')
elif 0.3 <= corr_abs < 0.5:
    print(f'A correlação de Pearson é fraca, com valor de {round(corr_depressao_sono, 2)}')
elif 0.5 <= corr_abs < 0.7:
    print(f'A correlação de Pearson é moderada, com valor de {round(corr_depressao_sono, 2)}')
elif corr_abs >= 0.7:
    print(f'A correlação de Pearson é forte, com valor de {round(corr_depressao_sono, 2)}')
```

```python
'''
H0: Não há diferença na média de bullying entre pessoas com saúde mental saudável e não saudável
    (as médias de bullying são iguais nos dois grupos)

Ha: Há diferença na média de bullying entre os dois grupos
    (pessoas com saúde mental não saudável sofreram mais/menos bullying)
'''

# Preparação dos dados

df['mental_health_history'] = df['mental_health_history'].replace({0: 'Não saudável', 1: 'Saudável'})

a = df.loc[df['mental_health_history'] == 'Não saudável', 'bullying']
b = df.loc[df['mental_health_history'] == 'Saudável', 'bullying']


# Verificação dos dados

print(f"Tamanho do grupo 'Não saudável': {len(a)}")
print(f"Tamanho do grupo 'Saudável': {len(b)}")
print(f"Média de bullying - Não saudável: {a.mean():.2f}")
print(f"Média de bullying - Saudável: {b.mean():.2f}")
print("-" * 30)

# Teste t

stat, p = ttest_ind(a, b)

print(f"Estatística T: {stat}")
print(f"Valor p: {p}")
print(f"-"*10)

# Interpretação

alpha = 0.05
if p < alpha:
    if a.mean() > b.mean():
        print("✅ Diferença significativa: Pessoas com saúde mental NÃO SAUDÁVEL sofreram MAIS bullying")
    else:
        print("✅ Diferença significativa: Pessoas com saúde mental SAUDÁVEL sofreram MAIS bullying")
    print("(Rejeita H0 - há evidências de diferença entre os grupos)")
else:
    print("❌ Não há diferença significativa na média de bullying entre os grupos")
    print("(Não rejeita H0)")
```
