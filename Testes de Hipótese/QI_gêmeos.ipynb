#Teste T

Hipótese nula - Não há diferença entre os QIs dos gêmeos, sejam eles criados pela família biológica ou famílias de adoção.

Hipótese alternativa - Há diferença entre os dois grupos.

```python
a = df['Foster']
b = df['Biological']

statistic, p_value = ttest_ind(a, b)
```

```python
print(f"Estatística T: {statistic}")
print(f"Valor p: {p_value}")
```

```python
alpha = 0.05
if p_value < alpha:
    print("Diferença estatisticamente significativa (rejeita H0)")
else:
    print("Não há diferença estatisticamente significativa (não rejeita H0)")
```
