informações: para X -> Y

**Support = frq(X,Y)/N**
---
Mostra a frequência em que os itens ocorrem juntos.

**Confidence = frq(X,Y)/frq(X)**
---
Mostra a probabilidade dos itens ocorrem juntos.

**Lift = support/supp(X)*supp(Y)**
---
Indica o grau de associação entre os itens.

```python
from pyECLAT import Example2, ECLAT
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import association_rules, apriori
```

#Analisando os dados

```python
dados = Example2().get()
```

```python
dados.head()
```

#ECLAT

```python
eclat = ECLAT(data=dados)
```

```python
indices, suporte = eclat.fit(min_support=0.08, verbose=False, min_combination=1)

print("Suporte:", suporte)
```

#APRIORI

```python
dados_esparsos = eclat.df_bin
```

```python
dados_esparsos
```

```python
# Gerando a associação
associacao = apriori(dados_esparsos, min_support=0.01, use_colnames=True)
```

```python
# Colocando em ordem dos mais frequentes
associacao.sort_values("support", ascending=False).head(15)
```

```python
# Criando as regras de associação
regras = association_rules(associacao, metric="confidence", num_itemsets=2, min_threshold=0.1)
```

```python
# Colocando em ordem de suporte, confiança ou grau de associação (Lift)
resultado = regras.sort_values("confidence", ascending=False)
```
